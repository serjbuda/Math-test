```python
import pandas as pd
import matplotlib.pyplot as plt

# Зчитуємо CSV файл та виводимо перші п'ять рядків
df = pd.read_csv("C:/Users/Сергій/Documents/Code/Current_task/Math-test/bestsellers with categories.csv")
print(df.head())

# Виводимо розмір датасету
print("Розмір датасету:", df.shape)
print("Кількість книг:", df.shape[0])

# Перейменовуємо стовпці з великих літер на snake_case
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

# Перевіряємо кількість пропусків в кожному стовпці
missing_values = df.isna().sum()
print("Пропуски в стовпцях:\n", missing_values)

# Унікальні значення в колонці genre
unique_genres = df['genre'].unique()
print("Унікальні жанри:", unique_genres)

# Розподіл цін
df['price'].plot(kind='hist')
plt.xlabel('Ціна')
plt.ylabel('Кількість книг')
plt.title('Розподіл цін')
plt.show()

# Статистики цін
max_price = df['price'].max()
min_price = df['price'].min()
mean_price = df['price'].mean()
median_price = df['price'].median()

print("Максимальна ціна:", max_price)
print("Мінімальна ціна:", min_price)
print("Середня ціна:", mean_price)
print("Медіанна ціна:", median_price)

# Найвищий рейтинг та кількість книг з таким рейтингом
max_rating = df['user_rating'].max()
count_max_rating = df[df['user_rating'] == max_rating].shape[0]

print("Найвищий рейтинг:", max_rating)
print("Кількість книг з найвищим рейтингом:", count_max_rating)

# Книга з найбільшою кількістю відгуків
most_reviewed_book = df[df['reviews'] == df['reviews'].max()]['name'].iloc[0]
print("Книга з найбільшою кількістю відгуків:", most_reviewed_book)

# Книга з найвищою ціною в 2015 році
max_price_2015 = df[df['year'] == 2015]['price'].max()
most_expensive_2015 = df[(df['year'] == 2015) & (df['price'] == max_price_2015)]['name'].iloc[0]
print("Найдорожча книга в 2015 році:", most_expensive_2015)

# Кількість книг жанру Fiction в 2010 році
count_fiction_2010 = df[(df['genre'] == 'Fiction') & (df['year'] == 2010)].shape[0]
print("Кількість книг жанру Fiction у 2010 році:", count_fiction_2010)

# Кількість книг з рейтингом 4.9 в 2010 та 2011 роках
count_rating_4_9 = df[(df['user_rating'] == 4.9) & ((df['year'] == 2010) | (df['year'] == 2011))].shape[0]
print("Кількість книг з рейтингом 4.9 в 2010 та 2011 роках:", count_rating_4_9)

# Сортування за зростанням ціни для книг 2015 року, які коштують менше 8 доларів
sorted_books_2015 = df[(df['year'] == 2015) & (df['price'] < 8)].sort_values(by='price')
last_book_sorted_2015 = sorted_books_2015.iloc[-1]['name']
print("Остання книга у відсортованому списку:", last_book_sorted_2015)
# Розподіл цін
df['price'].plot(kind='hist')
plt.xlabel('Ціна')
plt.ylabel('Кількість книг')
plt.title('Розподіл цін')
plt.show()
```
